```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("house_prediction_data.csv")
```

```python
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
x= data.drop(['Price'], axis=1)
y = data['Price']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(20,9))
```

```python
train_data.corr()
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['No. of Bedrooms']= np.log(train_data['No. of Bedrooms'] + 1)
train_data['MultipurposeRoom']= np.log(train_data['MultipurposeRoom'] + 1)
train_data['SwimmingPool']= np.log(train_data['SwimmingPool'] + 1)
train_data['CarParking']= np.log(train_data['CarParking'] + 1)
train_data['LiftAvailable']= np.log(train_data['LiftAvailable'] + 1)
```

```python
train_data.hist(figsize=(20,9))
```

```python
train_data.RainWaterHarvesting.value_counts()
```

```python
pd.get_dummies(train_data.RainWaterHarvesting)
```

```python
train_data = train_data.join(pd.get_dummies(train_data.RainWaterHarvesting)).drop(['RainWaterHarvesting'],axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(25,10))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="Area", y="ATM", data= train_data, hue="Price", palette="coolwarm")
```
