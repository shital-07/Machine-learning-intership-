```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

```

```python
wine= pd.read_csv("WineQT.csv")
```

```python
wine
```

```python
wine.info()
```

```python
from sklearn.model_selection import train_test_split
x= wine.drop(['quality'], axis=1)
y = wine['quality']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(20,9))
```

```python
train_data.corr()
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['fixed acidity']= np.log(train_data['fixed acidity'] + 1)
train_data['volatile acidity']= np.log(train_data['volatile acidity'] + 1)
train_data['citric acid']= np.log(train_data['citric acid'] + 1)
train_data['residual sugar']= np.log(train_data['residual sugar'] + 1)
train_data['chlorides']= np.log(train_data['chlorides'] + 1)
train_data['free sulfur dioxide']= np.log(train_data['free sulfur dioxide'] + 1)
train_data['density']= np.log(train_data['density'] + 1)
```

```python
train_data.hist(figsize=(20,9))
```

```python
train_data.sulphates.value_counts()
```

```python
pd.get_dummies(train_data.sulphates)
```

```python
train_data = train_data.join(pd.get_dummies(train_data.sulphates)).drop(['sulphates'],axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(50,15))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="pH", y="alcohol", data= train_data, hue="quality", palette="coolwarm")
```
